```python
df=dataset
df["Search Query"] = dataset["Search Query"].astype(str)
```

```python
# Rename the 'OldColumn' to 'NewColumn'
df.rename(columns={"Search Query": "text"}, inplace=True)
```

```python
df["text_lower"] = df["text"].str.lower()
df.head()
```

```python

PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    """custom function to remove the punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

df["text_wo_punct"] = df["text_lower"].apply(lambda text: remove_punctuation(text))
df.head()
```

```python
from nltk.corpus import stopwords
stop_words=stopwords.words('english')
```

```python
stop_words=sorted(stop_words)
# List of words to remove
words_to_remove = ["under", "above","after","before","until"]

# Remove the words from the list using list comprehension
stop_words= [word for word in stop_words if word not in words_to_remove]
```

```python
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

df["text_wo_stop"] = df["text_wo_punct"].apply(lambda text: remove_stopwords(text))
df.head()
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])

df["text_lemmatized"] = df["text_wo_stop"].apply(lambda text: lemmatize_words(text))
df.head()
```
